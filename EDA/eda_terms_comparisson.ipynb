```python
def compare_terms(filenames, path= "./raw_data/ICFES - SABER 11/"):
    '''
    Compare fields from different terms.
    input: 
        filenames: list(str) list of the names of the files from each term
        path: str location of the files 
    output: 
        df : dataframe where columns are the fields and rows are the terms,
             1 meaning that that specific term has the specific field, 
             0 otherwise. 
    '''
    all_terms = {}
    for file in file_names: 
        data = pd.read_csv(path+file, engine='python', sep = 'Â¬')
        term = file.split('.')[0]
        all_terms[term] = list(data.columns)
    del data
    terms_comparison = []; 
    for term in all_terms:
        fields = all_terms[term]
        term_dict = {}
        term_dict['SEMESTRE']= term.split('_')[1]
        for field in fields: 
            term_dict[field]= 1
        terms_comparison.append(term_dict)
    return pd.DataFrame(terms_comparison).fillna(0)
```
