```python
def value_counts_as_tuples(values, school):
    all_tuples = []
    for val in values.items(): 
        all_tuples.append(val)
    return all_tuples
```

```python
def normalize_name(txt1):
    if(pd.isna(txt1)):
        return ""
    clean_dict={
        'COL ': 'COLEGIO ',
        'FUND ': 'FUNDACION ',
        'BOGOTA':'BOGOTÁ',
        'BOYACA': 'BOYACA'
    }
    clean_name = txt1
    for abr in clean_dict.keys():
        clean_name = clean_name.replace(abr, clean_dict[abr])
    clean_name = clean_name.strip()
    return clean_name
```

```python
def make_normalization(txt1):
    try: 
        return normalize_name(txt1)
    except Exception as e:
        print(txt1)
        print(e)
        return ""
```

```python
def clean_term_schools(data):
    all_schools_ids = data['COLE_COD_DANE_SEDE'].drop_duplicates()
    score_fields = [ field for field in data.columns if re.search('_mean',field)]
    clean_schools = []
    count_all = len(all_schools_ids)
    for idx, school in enumerate(all_schools_ids):
        print(f"{idx}/{count_all}", end= '\r')
        all_samples = data[data['COLE_COD_DANE_SEDE']==school]

        ##CLENA OFICIAL SCHOOLS - KEEP ONLY NON OFFICIAL 
        school_nature = all_samples['COLE_NATURALEZA'].drop_duplicates()
        if(len(school_nature)>1):
            print(f'WARNING! School seat with DANE id {school} has several school nature: {school_nature.values}')
        if(school_nature.values[0] == 'OFICIAL' ):
            continue; 
        

        for school_shift in all_samples['COLE_JORNADA'].drop_duplicates().values:
            shift_samples = all_samples[all_samples['COLE_JORNADA']==school_shift]
            clean_school = {}


            ## TOTAL SAMPLES 
            clean_school['COLE_COD_DANE_SEDE']= school
            clean_school['TOTAL_ESTU'] = len(shift_samples)
            clean_school['COLE_NATURALEZA'] = school_nature.values[0]
            clean_school['COLE_JORNADA'] = school_shift


            ## NAMES NORMALIZATION 
            uniqueness_fields = ['COLE_NOMBRE_SEDE','COLE_MCPIO_UBICACION', 'COLE_DEPTO_UBICACION', 'COLE_NOMBRE_ESTABLECIMIENTO']
            for field in uniqueness_fields: 
                field_value = shift_samples[field].drop_duplicates()
                field_value = set([ make_normalization(i) for i in field_value])
                if(len(field_value)>1):
                    print(f'WARNING! school seat with DANE id {school} has {len(field_value)} {field}, wich are: \n {field_value}')
                clean_school[field] = field_value.pop()
            
            ## UNIQUENESS FIELDS 
            uniqueness_fields = ['COLE_COD_DANE_ESTABLECIMIENTO', 'COLE_GENERO', 'COLE_CALENDARIO', 'COLE_BILINGUE', 
            'COLE_CARACTER', 'COLE_SEDE_PRINCIPAL', 'COLE_AREA_UBICACION'
            ,'COLE_COD_MCPIO_UBICACION', 'COLE_COD_DEPTO_UBICACION']
            for field in uniqueness_fields: 
                field_value = shift_samples[field].drop_duplicates()
                if(len(field_value)>1):
                    print(f'WARNING! School seat with DANE id {school} has {len(field_value)} {field}, wich are: \n {field_value.values}')
                if(len(field_value) == 0):
                    print(f'WARNING! School seat with DANE id {school} has no values for field: {field}')

                clean_school[field] = field_value.values[0]
            
            ## FIELD WITH MULTIPLE VALUES FOR SAME SCHOOL SEAT
            multiple_value_fields = ['COLE_CODIGO_ICFES']
            for field in multiple_value_fields: 
                clean_school[field] = value_counts_as_tuples(shift_samples[field].value_counts(), school)

            ## ICFES SCORES RELATED FIELDS
            for field in score_fields:
                desc_stat = shift_samples[field].describe()
                for desc in desc_stat.index:
                    if (desc == 'count'):
                        continue
                    key_name = f'{field}_{desc}'
                    clean_school[key_name] = desc_stat[desc]

            ## SAVE CLEAN VALUE
            clean_schools.append(clean_school)

    return clean_schools
```

### Comparisson functions

```python
def normalize_name(txt1, clean_dict={}, debug=False):
    if(pd.isna(txt1)):
        return ""
    clean_name = txt1.lower()
    clean_name = re.sub('\(ant.*?\)', ' ', clean_name)
    remove_accents = clean_name.maketrans("áéíóúüñ", "aeiouun")
    clean_name = clean_name.translate(remove_accents)
    clean_name = re.sub('^i\.?e\.? ', 'institucion educativa ', clean_name)  
    clean_name = clean_name.translate(str.maketrans('', '', string.punctuation))
    for abr in clean_dict.keys():
        clean_name = clean_name.replace(abr, clean_dict[abr])
    clean_name = re.sub(' +', ' ', clean_name) 
    clean_name = clean_name.strip()
    if(debug) : print(f"'{clean_name}'", end=":")
    return clean_name
    
def compare_data_values(txt1, txt2, clean_dict,  debug = False): 
    try:
        return normalize_name(txt1, clean_dict , debug) == normalize_name(txt2, clean_dict , debug)
    except Exception as e:
        print(f'UNABLE TO NORMALIZE {txt1} and {txt2}')
        print(e)
        return False
        
def compare_int_values(num1, num2): 
    if (pd.isna(num1) | pd.isna(num2)):
        return False
    try:
        return int(num1) == int(num2)
    except Exception as e:
        print(f'UNABLE TO COMPARE {num1} and {num2}')
        print(e)
        return False

def compare_list_value(val, arr, clean_dict): 
    if (pd.isna(val) | pd.isna(arr)):
        return False
    try:
        arr = arr.split(',')
        arr = [normalize_name(i, clean_dict) for i in arr]
        return normalize_name(val) in arr
    except Exception as e:
        print(f'UNABLE TO COMPARE {val} in {arr}')
        print(e)
        return False
```

```python
sineb_data = pd.read_csv(f'clean_schools/colegios_establecimientos.csv', sep = ";", encoding='latin', dtype={'codigo departamento': np.int32})
school_data = pd.read_csv(f'clean_schools/clean_schools_20172.csv')
sineb_data.rename(normalize_name, axis='columns',inplace=True)
rich_data = school_data.merge(sineb_data, how= "left", left_on=id['icfes'], right_on=id['sineb'])
rich_data.head(2)
clean_department = {
        'bogota dc': 'bogota',
        'valle del cauca': 'valle',
        'norte de santander': 'norte santander',
        'archipielago de san andres providencia y santa catalina': 'san andres'
}
clean_municipality = {
        'bogota dc': 'bogota',
        'dos quebradas':'dosquebradas',
        'san andres de tumaco':'tumaco',
        'santiago de tolu':'tolu',
        'ciudad bolivar':'bolivar',
        'villa de san diego de ubate':'ubate',
        'guadalajara de buga':'buga',
        'san sebastian de mariquita':'mariquita',
        'san luis de since':'since',
        'el santuario':'santuario',
        'san andres de sotavento':'san andres sotavento',
        'el carmen de chucuri':'el carmen',
        'san pedro de los milagros':'san pedro',
        'don matias':'donmatias',
        'el carmen de viboral':'carmen de viboral',
        'ariguani el dificil':'ariguani',
        'la paz robles':'la paz',
        'patiael bordo':'patia',
        'calima darien':'calima',
        ' del ': '',
        ' de ': '',

}
clean_shift = {
    'fin de semana':'sabatina','nocturna':'noche',
}
to_compare = [
    (departament['sineb'], departament['icfes'], clean_department, "text"),
    (departmend_code['sineb'], departmend_code['icfes'], {}, "num"),
    (calendar['sineb'], calendar['icfes'], {}, "other"),
    (municipality['sineb'], municipality['icfes'], clean_municipality, "text"), 
    (municipality_code['sineb'], municipality_code['icfes'], {}, "num"),
    (gender['sineb'], gender['icfes'], {}, "text"),
    (shift['sineb'], shift['icfes'], {}, "list"), 
]
for sineb, icfes, clean_dict, is_text in to_compare: 
    if(is_text == "text"):
        aux_df = rich_data[(pd.notna(rich_data[id['sineb']]) & 
                rich_data.apply(lambda x: not(compare_data_values(x[sineb], x[icfes],clean_dict)), axis=1))][[id['icfes'],sineb, icfes]]
    elif(is_text == "num"): 
        aux_df = rich_data[(pd.notna(rich_data[id['sineb']]) & 
                rich_data.apply(lambda x: not(compare_int_values(x[sineb],x[icfes])), axis=1))][[id['icfes'],sineb, icfes]]
    elif(is_text == "list"): 
        aux_df = rich_data[(pd.notna(rich_data[id['sineb']]) & 
                rich_data.apply(lambda x: not(compare_list_value(x[icfes],x[sineb],clean_shift)), axis=1))][[id['icfes'],sineb, icfes]]
    else: 
        aux_df = rich_data[pd.notna(rich_data[id['sineb']]) & 
                rich_data.apply(lambda x: x[sineb] != x[icfes], axis=1)][[id['icfes'],sineb, icfes]]        
    if(len(aux_df) > 0): 
        print(f'WARNING: NON MATCHING {icfes}. SAMPLE VALUES:\n {aux_df.drop_duplicates(subset=[sineb, icfes])[[sineb,icfes]]}')
```
