```python
import pandas as pd
import re
```

# The data in the terms

```python
field_resummen = pd.read_csv('./field_resummne_save.csv', engine='python')
school_fields = [ field for field in field_resummen.columns if re.search('COLE',field)]
field_resummen[school_fields].transpose()
```

```python
terms_count = len(field_resummen['SEMESTRE'].drop_duplicates())
common_fields = [field for field in school_fields if field_resummen[field].sum() == terms_count]
print(common_fields)
del field_resummen, school_fields, terms_count
```

# Focusing in one term - allow parametrization

```python
data = pd.read_csv('./raw_data/ICFES - SABER 11/SB11_20212.txt', engine='python', sep = '¬')
to_drop = [field for field in data.columns if field not in common_fields]
data.drop(to_drop, axis=1, inplace=True)
del data
```
