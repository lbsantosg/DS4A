# Focusing in one term - allow parametrization

```python
def value_counts_as_tuples(values, school):
    all_tuples = []
    for val in values.items(): 
        all_tuples.append(val)
    return all_tuples
```

```python
def clean_term_schools(data):
    all_schools_ids = data['COLE_COD_DANE_SEDE'].drop_duplicates()
    score_fields = [ field for field in data.columns if re.search('PUNT',field)]
    clean_schools = []
    count_all = len(all_schools_ids)
    for idx, school in enumerate(all_schools_ids):
        print(f"{idx}/{count_all}", end= '\r')
        all_samples = data[data['COLE_COD_DANE_SEDE']==school]

        ##CLENA OFICIAL SCHOOLS - KEEP ONLY NON OFFICIAL 
        school_nature = all_samples['COLE_NATURALEZA'].drop_duplicates()
        if(len(school_nature)>1):
            print(f'WARNING! School seat with DANE id {school} has several school school nature: {school_nature.values}')
        if(school_nature.values[0] == 'OFICIAL' ):
            continue; 

        for school_shift in all_samples['COLE_JORNADA'].drop_duplicates().values:
            shift_samples = all_samples[all_samples['COLE_JORNADA']==school_shift]
            clean_school = {}


            ## TOTAL SAMPLES 
            clean_school['COLE_COD_DANE_SEDE']= school
            clean_school['TOTAL_ESTU'] = len(shift_samples)
            clean_school['COLE_NATURALEZA'] = school_nature.values[0]
            clean_school['COLE_JORNADA'] = school_shift

            ## UNIQUENESS FIELDS 
            uniqueness_fields = ['COLE_COD_DANE_ESTABLECIMIENTO', 'COLE_NOMBRE_ESTABLECIMIENTO', 'COLE_GENERO', \
            'COLE_CALENDARIO', 'COLE_BILINGUE', 'COLE_CARACTER', 'COLE_NOMBRE_SEDE', 'COLE_SEDE_PRINCIPAL', 'COLE_AREA_UBICACION', \
            'COLE_COD_MCPIO_UBICACION', 'COLE_MCPIO_UBICACION', 'COLE_COD_DEPTO_UBICACION', 'COLE_DEPTO_UBICACION']
            for field in uniqueness_fields: 
                field_value = shift_samples[field].drop_duplicates()
                if(len(field_value)>1):
                    print(f'WARNING! School seat with DANE id {school} has {len(field_value)} {field}, wich are: \n {field_value.values}')
                if(len(field_value) == 0):
                    print(f'WARNING! School seat with DANE id {school} has no values for field: {field}')

                clean_school[field] = field_value.values[0]
            
            ## FIELD WITH MULTIPLE VALUES FOR SAME SCHOOL SEAT
            multiple_value_fields = ['COLE_CODIGO_ICFES']
            for field in multiple_value_fields: 
                clean_school[field] = value_counts_as_tuples(shift_samples[field].value_counts(), school)

            ## ICFES SCORES RELATED FIELDS
            for field in score_fields:
                desc_stat = shift_samples[field].describe()
                for desc in desc_stat.index:
                    if (desc == 'count'):
                        continue
                    key_name = f'{field}_{desc}'
                    clean_school[key_name] = desc_stat[desc]

            ## SAVE CLEAN VALUE
            clean_schools.append(clean_school)
    del shift_samples, all_schools_ids, clean_school, field, field_value, multiple_value_fields, school, uniqueness_fields
    return clean_schools
```

```python
%%time
file_names = [
    "SB11_20212.txt",
    "SB11_20211.txt",
    "SB11_20202.txt",
    "SB11_20201.txt",
    "SB11_20192.TXT",
    "SB11_20191.TXT",
    "SB11_20182.TXT",
    "SB11_20181.TXT",
    "SB11_20172.TXT",
    "SB11_20171.TXT",
    # "SB11_20162.TXT",
    # "SB11_20161.TXT",
    # "SB11_20152.txt",
    # "SB11_20151.txt",
]
common_fields= ['COLE_CODIGO_ICFES', 'COLE_COD_DANE_ESTABLECIMIENTO', 'COLE_NOMBRE_ESTABLECIMIENTO', 'COLE_GENERO', 'COLE_NATURALEZA', 'COLE_CALENDARIO', 'COLE_BILINGUE', 'COLE_CARACTER', 'COLE_COD_DANE_SEDE', 'COLE_NOMBRE_SEDE', 'COLE_SEDE_PRINCIPAL', 'COLE_AREA_UBICACION', 'COLE_JORNADA', 'COLE_COD_MCPIO_UBICACION', 'COLE_MCPIO_UBICACION', 'COLE_COD_DEPTO_UBICACION', 'COLE_DEPTO_UBICACION', 'PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS', 'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES', 'PUNT_GLOBAL']
for name in file_names[:]: 
    term = name.split("_")[1].split(".")[0]
    print(f'Working on term: {term}')
    data = pd.read_csv(f'./raw_data/ICFES - SABER 11/{name}', engine='python', sep = '¬')
    to_drop = [field for field in data.columns if field not in common_fields]
    data.drop(to_drop, axis=1, inplace=True)
    print(f'Start schools cleaning...: {term}')
    clean_schools = pd.DataFrame(clean_term_schools(data))
    clean_schools.to_csv(f"clean_schools_{term}.csv")
    print(f'Done on term: {term}')
```
