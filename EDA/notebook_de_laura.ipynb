## 3. Estadistica descriptiva (variables continuas) y BoxPlot Puntaje (variables discretas)

Las variables continuas a considerar son siguientes:

* Lectura Critica
* Matemáticas
* Ciencias Naturales
* Ciencias Sociales y Ciudadanas
* Puntaje Global


````python
ICFES20192work                         =  ICFES20192

# Calculo edad al presentar la prueba Saber 11

Saber_DAY                              =  pd.to_datetime("08/11/2019", errors="coerce")                               # Día aplicación prueba Saber
ICFES20192work["ESTU_FECHANACIMIENTO"] =  pd.to_datetime(ICFES20192work["ESTU_FECHANACIMIENTO"], errors="coerce")     # Trasnformación fecha de nacimiento
ICFES20192work["EDAD"]                 =  (Saber_DAY - ICFES20192work["ESTU_FECHANACIMIENTO"])/np.timedelta64(1,'Y')  # cálculo número de años

ICFES20192work=ICFES20192work[ICFES20192work.EDAD>=15] # filtro personas con 15 años o más

ICFES20192work[["PUNT_LECTURA_CRITICA","PUNT_MATEMATICAS","PUNT_C_NATURALES","PUNT_SOCIALES_CIUDADANAS","PUNT_INGLES","PUNT_GLOBAL","EDAD"]].describe()
````
