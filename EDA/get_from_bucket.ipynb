```python
def clean_from_zip(raw_data_path = "./raw_data/ICFES - SABER 11/" , csv_files= "./raw_data/ICFES_CSV/" ):
    """
    Takes files just as they came after extraction from .zip and converts them
    into a python redable cvs files. 
    input: 
    'raw_data_path' : path where the .txt files are located. All files should be in the same dir. 
    'csv_files' : path where the .cvs will be located after conversion

    """
    file_names = [
    "SB11_20212.txt",
    "SB11_20211.txt",
    "SB11_20202.txt",
    "SB11_20201.txt",
    "SB11_20192.TXT",
    "SB11_20191.TXT",
    "SB11_20182.TXT",
    "SB11_20181.TXT",
    "SB11_20172.TXT",
    "SB11_20171.TXT",
    "SB11_20162.TXT",
    "SB11_20161.TXT",
    "SB11_20152.txt",
    "SB11_20151.txt",
    ]
    if (not os.path.exists(csv_files)):
        os.makedirs(csv_files)

    for f in file_names : 
        term = f.split('.')[0]; 
        file = open( raw_data_path + f);
        text =file.read();
        csv_file = open(csv_files + term + ".csv" ,"w")
        csv_file.write(text.replace("Â¬", ","));
```

```python
def get_data_from_aws(file_names, client, bucket_name = "proyect-team166", raw_data_path = "./raw_data/ICFES_AWS_CSV/"):
    
    if (not os.path.exists(raw_data_path)):
        os.makedirs(raw_data_path)

    for file in file_names:
        object_key = "preprocess/ICFES_CSV/"+ file
        csv_obj = client.get_object(Bucket=bucket_name, Key=object_key);
        body = csv_obj['Body']
        csv_string = body.read().decode('utf-8')
        text_file = open(raw_data_path + file,"w")
        text_file.write(csv_string)
        text_file.close()
```

```python
aws_id = os.getenv('AWS_ID')
aws_secret = os.getenv('AWS_SECRET')
client = boto3.client('s3', aws_access_key_id = aws_id, aws_secret_access_key = aws_secret);
file_names = [
    "SB11_20212.csv",
    "SB11_20211.csv",
    # "SB11_20202.csv",
    # "SB11_20201.csv",
    # "SB11_20192.csv",
    # "SB11_20191.csv",
    # "SB11_20182.csv",
    # "SB11_20181.csv",
    # "SB11_20172.csv",
    # "SB11_20171.csv",
    # "SB11_20162.csv",
    # "SB11_20161.csv",
    # "SB11_20152.csv",
    # "SB11_20151.csv",
]
get_data_from_aws(file_names, client)
```
